```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fungsi Keanggotaan
class MembershipFunction:
    def __init__(self, name, points):
        self.name = name
        self.points = points

    def calculate(self, x):
        if len(self.points) == 3:  # Segitiga
            a, b, c = self.points
            if a <= x < b:
                return (x - a) / (b - a)
            elif b <= x < c:
                return (c - x) / (c - b)
            else:
                return 0.0
        elif len(self.points) == 4:  # Trapesium
            a, b, c, d = self.points
            if a <= x < b:
                return (x - a) / (b - a)
            elif b <= x <= c:
                return 1.0
            elif c < x <= d:
                return (d - x) / (d - c)
            else:
                return 0.0

    def plot(self, x_range):
        y_values = [self.calculate(x) for x in x_range]
        plt.plot(x_range, y_values, label=self.name)
```

```python
# Fuzzifikasi
class Fuzzification:
    def __init__(self, membership_functions):
        self.membership_functions = membership_functions

    def fuzzify(self, value):
        fuzzy_values = {}
        for mf in self.membership_functions:
            fuzzy_values[mf.name] = mf.calculate(value)
        return fuzzy_values
```

```python
# Evaluasi Aturan Fuzzy
class RuleEvaluation:
    def __init__(self, rules):
        self.rules = rules

    def evaluate(self, temp_fuzzy, hum_fuzzy):
        rule_results = []
        for rule in self.rules:
            temp_level, hum_level, output = rule
            rule_strength = min(temp_fuzzy[temp_level], hum_fuzzy[hum_level])
            rule_results.append((rule_strength, output))
        return rule_results
```

```python
# Defuzzifikasi
class Defuzzification:
    def __init__(self, output_ranges):
        self.output_ranges = output_ranges

    def defuzzify(self, rules):
        numerator = 0.0
        denominator = 0.0
        for strength, output_name in rules:
            crisp_value = self.output_ranges[output_name]
            numerator += strength * crisp_value
            denominator += strength
        return numerator / denominator if denominator != 0 else 0.0
```

```python
# Sistem Inferensi Fuzzy
class FuzzySystem:
    def __init__(self, temp_mfs, hum_mfs, rules, output_ranges):
        self.temp_fuzz = Fuzzification(temp_mfs)
        self.hum_fuzz = Fuzzification(hum_mfs)
        self.rule_eval = RuleEvaluation(rules)
        self.defuzz = Defuzzification(output_ranges)

    def calculate(self, temp_value, hum_value):
        temp_fuzzy = self.temp_fuzz.fuzzify(temp_value)
        hum_fuzzy = self.hum_fuzz.fuzzify(hum_value)
        rule_results = self.rule_eval.evaluate(temp_fuzzy, hum_fuzzy)
        return self.defuzz.defuzzify(rule_results)
```

```python
# Definisi Fungsi Keanggotaan
temp_mfs = [
    MembershipFunction("Cold", [0, 0, 20]),
    MembershipFunction("Moderate", [15, 25, 35]),
    MembershipFunction("Hot", [30, 40, 40]),
]

hum_mfs = [
    MembershipFunction("Dry", [0, 0, 50]),
    MembershipFunction("Normal", [25, 50, 75]),
    MembershipFunction("Humid", [50, 100, 100]),
]
```

```python
# Aturan Fuzzy
rules = [
    ("Cold", "Dry", "Slow"),
    ("Moderate", "Normal", "Medium"),
    ("Hot", "Humid", "Fast"),
]
```

```python
# Output Ranges (Crisp)
output_ranges = {
    "Slow": 2.5,
    "Medium": 5.0,
    "Fast": 7.5,
}
```

```python
# Inisialisasi Sistem
fuzzy_system = FuzzySystem(temp_mfs, hum_mfs, rules, output_ranges)
```

```python
# Input
temp_input = 28  # Suhu (°C)
hum_input = 60   # Kelembapan (%)
```

```python
# Hitung Output
fan_speed = fuzzy_system.calculate(temp_input, hum_input)
```

```python
# Output
print(f"Kecepatan Kipas: {fan_speed:.2f}")
```

```python
# Visualisasi Fungsi Keanggotaan
def plot_membership_functions():
    x_temp = np.linspace(0, 40, 1000)
    x_hum = np.linspace(0, 100, 1000)
    x_speed = np.linspace(0, 10, 1000)

    plt.figure(figsize=(10, 4))
    for mf in temp_mfs:
        mf.plot(x_temp)
    plt.title("Fungsi Keanggotaan Suhu")
    plt.xlabel("Suhu (°C)")
    plt.ylabel("Derajat Keanggotaan")
    plt.legend()
    plt.grid()
    plt.show()

    plt.figure(figsize=(10, 4))
    for mf in hum_mfs:
        mf.plot(x_hum)
    plt.title("Fungsi Keanggotaan Kelembapan")
    plt.xlabel("Kelembapan (%)")
    plt.ylabel("Derajat Keanggotaan")
    plt.legend()
    plt.grid()
    plt.show()

    plt.figure(figsize=(10, 4))
    slow = MembershipFunction("Slow", [0, 0, 5])
    medium = MembershipFunction("Medium", [3, 5, 7])
    fast = MembershipFunction("Fast", [5, 10, 10])
    for mf in [slow, medium, fast]:
        mf.plot(x_speed)
    plt.title("Fungsi Keanggotaan Kecepatan Kipas")
    plt.xlabel("Kecepatan Kipas")
    plt.ylabel("Derajat Keanggotaan")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
# Menjalankan Visualisasi
plot_membership_functions()
```
